```python
results = pd.read_csv("./figures/advanced_models/xgb_hparam_search_results.csv")
```

```python
hparams = [
    'n_estimators', 'max_depth', 'colsample_bytree', 'subsample', 'learning_rate', 'gamma'
]

hparam_cols = ['param_' + hparam for hparam in hparams]
```

## One-Dimensional Scatters

```python
fig, axes = plt.subplots(nrows=len(hparam_cols), ncols=1, figsize=(10,20));

for i, hparam in enumerate(hparam_cols):
    axes[i].scatter(x=results[hparam], y=results['mean_train_score'], label=hparams[i])
    axes[i].set_xlabel(hparams[i])
    
plt.style.use('seaborn-white');
plt.tight_layout();
```

```python
fig, axes = plt.subplots(nrows=len(hparam_cols), ncols=1, figsize=(10,20));

results = results[results['param_max_depth']==1]

for i, hparam in enumerate(hparam_cols):
    axes[i].scatter(x=results[hparam], y=results['mean_train_score'], label=hparams[i])
    axes[i].set_xlabel(hparams[i])
    
plt.style.use('seaborn-white');
plt.tight_layout();
```
