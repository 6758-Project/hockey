```python
import os
import pandas as pd
from comet_ml import Experiment
```

```python
game_id = 2017021065
file_path = os.path.join("../data/cleaned/", str(game_id) + '.csv')
events_df = pd.read_csv(file_path, index_col=0)
events_df.columns
```

```python
saved_cols = ["game_id", "home_team", "away_team", "type", "period", "distance_from_net", "angle", "secondary_type", "is_empty_net",
              "coordinate_x", "coordinate_y", "game_sec", "prev_event_type", "prev_event_x_coord", 
              "prev_event_y_coord", "prev_event_time_diff", "distance_from_prev_event", 
              "is_rebound", "rebound_angle", "speed", 'friendly_skaters', 'opposing_skaters',
       'power_play_time_elapsed']
events_df[saved_cols].head()
```

```python
# save the dataframe with all the required features in Q4 in milestone-2
my_comet_apt_key = os.environ.get("COMET_API_KEY")
```

```python
EXP_KWARGS = {
    'project_name': 'ift6758-hockey',
    'workspace': "tim-k-lee",
    'auto_param_logging': False
}

experiment = Experiment(api_key=my_comet_apt_key, **EXP_KWARGS)

experiment.set_name('feature_engineering_dataframe')

experiment.log_dataframe_profile(
    events_df[saved_cols],
    name='wpg_v_wsh_2017021065', # keep this name
    dataframe_format='csv')

experiment.end()
```
