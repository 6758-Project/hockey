```python
import os
import sys
import pandas as pd
sys.path.insert(0,'../src/data/')
from team_data import  get_all_teams 
```

```python
# read the cleaned data files 
sub_seasons = ['regular', 'postseason']
years = [2016, 2017, 2018, 2019, 2020]
teams_lst = get_all_teams(data_dir="../data/raw")

```

```python
# check the median of the x-coordinate to see where did every team start the game (left or right)
for year in years:
    for sub_season in sub_seasons:
        game_file_str = str(year) + '-' + sub_season + '.csv'
        games_df = pd.read_csv(os.path.join("../data/games/", game_file_str))
        
        median_shot_coordinates = games_df[(games_df['period']==1)]\
                                    .groupby(['shooter_team_name', 'home_team', 'period'])\
                                    [['coordinate_x', 'coordinate_y']].median().reset_index()
        
```
